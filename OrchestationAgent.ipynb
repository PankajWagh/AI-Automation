agent orchestration with routines and handoffs. 
- Agents: Independent units (often wrapping an LLM) that specialize in tasks (retrieval, summarization, evaluation, etc.).
- Routines: Predefined sequences of steps an agent follows (like a workflow pipeline).
- Handoffs: Passing context or results from one agent to another, ensuring smooth transitions and modularity.

- Retriever → returns JSON with documents
- Summarizer → consumes documents and outputs summary
- Evaluator → consumes summary and outputs score

 Typical Architecture in Python + OpenA

```python
from openai import OpenAI
import asyncio

client = OpenAI()

# Define a generic agent wrapper
class Agent:
    def __init__(self, role, instructions):
        self.role = role
        self.instructions = instructions

    async def run(self, input_text):
        response = client.chat.completions.create(
            model="gpt-4-turbo",
            messages=[
                {"role": "system", "content": self.instructions},
                {"role": "user", "content": input_text}
            ]
        )
        return response.choices[0].message.content

# Example agents
retriever = Agent("Retriever", "You are a retrieval agent. Find relevant context.")
summarizer = Agent("Summarizer", "You are a summarizer. Condense input clearly.")
evaluator = Agent("Evaluator", "You are an evaluator. Judge relevance and accuracy.")
```

 Orchestration with Routines & Handoff

```python
# Define orchestration as async
async def orchestrate(query):
    print("\n retriever:")
    context = await retriever.run(query)
    print(context)
    print("\n summarizer:")
    summary = await summarizer.run(context)
    print(summary)
    print("\n evaluator:")
    evaluation = await evaluator.run(summary)
    print(evaluator)
    print("\n*********************\n")
    return {"context": context, "summary": summary, "evaluation": evaluation}

# Call with await
result = await orchestrate("Explain LangChain agent handoffs")
print(result)
```
