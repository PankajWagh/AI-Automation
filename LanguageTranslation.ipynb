```python
from openai import OpenAI
client = OpenAI()
```

Define a Translation Function

```python
def translate_text(text, target_language="French"):
    response = client.chat.completions.create(
        model="gpt-4",
        messages=[
            {"role": "system", "content": f"You are a translator that converts text into {target_language}."},
            {"role": "user", "content": text}
        ]
    )
    return response.choices[0].message.content
```

Use in a Notebook Cell

```python
english_text = "The training process of a PyTorch model involves forward and backward passes."
translated = translate_text(english_text, target_language="Spanish")
print(translated)
```

Extend for Multiple Languages

```python
languages = ["French", "German", "Hindi"]

for lang in languages:
    print(f"\n--- {lang} ---")
    print(translate_text("Hello, how are you?", target_language=lang))
```
