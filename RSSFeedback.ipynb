```python
import feedparser

rss_url = "https://news.google.com/rss"
feed = feedparser.parse(rss_url)

articles = []
for entry in feed.entries[:10]:  # limit for demo
    articles.append({
        "title": entry.title,
        "summary": entry.summary,
        "link": entry.link
    })

```

```python
from openai import OpenAI
client = OpenAI()

summaries = []
for article in articles:
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role":"user","content":f"Summarize this news in 2 sentences:\n{article['title']} - {article['summary']}"}]
    )
    summaries.append({
        "title": article["title"],
        "summary": response.choices[0].message.content,
        "link": article["link"]
    })
print(summaries);
```
