##temperature
Controls the randomness of AI responses by adjusting the probability distribution. This parameter determines how predictable or creative the model's output will be. Lower values make the model more focused and deterministic, while higher values increase diversity and creativity. It's one of the most important parameters for controlling response quality

Low (0.0 – 0.3): "Factual Mode"
Perfect for data extraction, code generation, and fact-based tasks.
Produces consistent, reliable outputs.

High (0.8 – 1.5): "Creative Mode"
Ideal for brainstorming, storytelling, and content creation. Generates
diverse and imaginative responses.


```python
from openai import OpenAI
import os
val = input("Enter your search value: ")
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
print("+++++++++++++++++++++++++++++++++++++++temparature 0 +++++++++++++++++++++++++++++++++++++++++")
try:
    response = client.responses.create(
        input=val,
        model="gpt-4.1",
        temperature=0
    )
    print(response.output_text)

except Exception as e:
    print(f"An error occurred: {e}")

print("+++++++++++++++++++++++++++++++++++++++temparature 0.7 +++++++++++++++++++++++++++++++++++++++++")
try:
    response = client.responses.create(
        input=val,
        model="gpt-4.1",
        temperature=0.7
    )
    print(response.output_text)

print("+++++++++++++++++++++++++++++++++++++++temparature 0.1 +++++++++++++++++++++++++++++++++++++++++")
try:
    response = client.responses.create(
        input=val,
        model="gpt-4.1",
        temperature=1.0
    )
    print(response.output_text)

except Exception as e:
    print(f"An error occurred: {e}")
    response = client.responses.create(
        input=val,
        model="gpt-4.1",
         tools=[{"type": "web_search"}],
    )
    print(response.output_text)
except Exception as e:
    print(f"An error occurred: {e}")
```

Filters the model's vocabulary based on cumulative probability, creating a focused token pool. Also known as nucleus sampling, this
technique selects from the smallest set of tokens whose probabilities add up to P. It dynamically adjusts the vocabulary size based on
confidence, making responses more coherent than pure random sampling.
Lower values (e.g., 0.3–0.5): More focused, deterministic outputs.
Higher values (e.g., 0.9–1): More diverse, creative outputs.

```python
import os
from openai import OpenAI

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

english_text = "What is 12*15 and the capital of France?"

response = client.responses.create(
    model="gpt-4.1",
    input=f"Translate to marathi: {english_text}",
    top_p=0.9   # controls nucleus sampling
)

print(response.output[0].content[0].text)
```

```python
def tokenize_unique(line: str):
    tokens = line.strip().split()
    unique_tokens = []
    seen = set()
    for word in tokens:
        if word not in seen:
            unique_tokens.append(word)
            seen.add(word)
    return unique_tokens
```

```python
from openai import OpenAI
client = OpenAI()

line = "apple banana apple orange banana grape"
tokens = tokenize_unique(line)

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": "You are a helpful tokenizer assistant."},
        {"role": "user", "content": f"Tokens: {tokens}"}
    ]
)

print(response.choices[0].message.content)
```

Discourages the model from repeating itself by penalizing tokens that have already appeared. These penalties range from -2.0 to 2.0, where positive values reduce repetition and negative values increase it. They're crucial for preventing the model from getting stuck in loops and for encouraging more diverse, natural-sounding outputs.

```python
prompt = "Write a short poem about simla in marathi."
```

##Types of Penalties
##frequency_penalty
    - Reduces the likelihood of the model repeating the same token (word/phrase).
    - Higher values discourage repetition.
    - Range: -2.0 to 2.0.
##presence_penalty
    - Encourages the model to introduce new tokens that haven’t appeared yet.
    - Useful when you want variety or exploration.
    - Range: -2.0 to 2.0.
##maxlength
Sets a hard limit on the total length of the AI's response, measured in tokens. This parameter is crucial for API cost management and preventing runaway generation. For example, setting max_tokens to 100 means the model will stop after approximately 100 tokens, even if incomplete. Use 500-1000 for detailed responses, 50-150 for concise answers

```python
response_no_penalty = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": prompt}],
    max_tokens=100,
    temperature=0.7,
    frequency_penalty=0.0,  # no penalty
    presence_penalty=0.0    # no penalty
)

print("=== Without Penalties ===")
print(response_no_penalty.choices[0].message.content)
```

```python
response_with_penalty = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": prompt}],
    max_tokens=100,
    temperature=0.7,
    frequency_penalty=0.8,  # discourage repetition
    presence_penalty=0.6    # encourage new words
)

print("=== With Penalties ===")
print(response_with_penalty.choices[0].message.content)
```
