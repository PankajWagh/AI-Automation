```python
from openai import OpenAI
import os
val = input("Enter your search value: ")
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
print("+++++++++++++++++++++++++++++++++++++++temparature 0 +++++++++++++++++++++++++++++++++++++++++")
try:
    response = client.responses.create(
        input=val,
        model="gpt-4.1",
        temperature=0
    )
    print(response.output_text)

except Exception as e:
    print(f"An error occurred: {e}")

print("+++++++++++++++++++++++++++++++++++++++temparature 0.7 +++++++++++++++++++++++++++++++++++++++++")
try:
    response = client.responses.create(
        input=val,
        model="gpt-4.1",
        temperature=0.7
    )
    print(response.output_text)

except Exception as e:
    print(f"An error occurred: {e}")
print("+++++++++++++++++++++++++++++++++++++++temparature 1.0 +++++++++++++++++++++++++++++++++++++++++")
try:
    response = client.responses.create(
        input=val,
        model="gpt-4.1",
        temperature=1
    )
    print(response.output_text)
except Exception as e:
    print(f"An error occurred: {e}")
print("+++++++++++++++++++++++++++++++++++++++Web Search+++++++++++++++++++++++++++++++++++++++++")
try:
    response = client.responses.create(
        input=val,
        model="gpt-4.1",
         tools=[{"type": "web_search"}],
    )
    print(response.output_text)
except Exception as e:
    print(f"An error occurred: {e}")
```
