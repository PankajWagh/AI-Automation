Define Action Functions

```python
def translate_text(text, target_language="French"):
    response = client.chat.completions.create(
        model="gpt-4",
        messages=[
            {"role": "system", "content": f"You are a translator that converts text into {target_language}."},
            {"role": "user", "content": text}
        ]
    )
    return response.choices[0].message.content

def summarize_text(text):
    response = client.chat.completions.create(
        model="gpt-4",
        messages=[
            {"role": "system", "content": "You are a summarizer that condenses text."},
            {"role": "user", "content": text}
        ]
    )
    return response.choices[0].message.content

def sentiment_analysis(text):
    response = client.chat.completions.create(
        model="gpt-4",
        messages=[
            {"role": "system", "content": "You are a sentiment analyzer that classifies text as positive, negative, or neutral."},
            {"role": "user", "content": text}
        ]
    )
    return response.choices[0].message.content

def extract_keywords(text):
    response = client.chat.completions.create(
        model="gpt-4",
        messages=[
            {"role": "system", "content": "You are a keyword extractor that identifies key terms from text."},
            {"role": "user", "content": text}
        ]
    )
    return response.choices[0].message.content
```

Interactive Multi‑Action UI

```python
# Dropdown for action selection
action_dropdown = widgets.Dropdown(
    options=["Translate", "Summarize", "Sentiment", "Keywords"],
    value="Translate",
    description="Action:"
)

# Text input
text_input = widgets.Textarea(
    value="Hello, how are you?",
    description="Input Text:",
    layout=widgets.Layout(width="600px", height="100px")
)

# Extra input for translation target language
lang_input = widgets.Text(
    value="French",
    description="Target Lang:"
)

# Output area
output_area = widgets.Output()

# Button
run_button = widgets.Button(description="Run Action")

def run_action(b):
    output_area.clear_output()
    with output_area:
        text = text_input.value
        action = action_dropdown.value
        if action == "Translate":
            print(translate_text(text, target_language=lang_input.value))
        elif action == "Summarize":
            print(summarize_text(text))
        elif action == "Sentiment":
            print(sentiment_analysis(text))
        elif action == "Keywords":
            print(extract_keywords(text))

run_button.on_click(run_action)

# Display widgets
display(action_dropdown, text_input, lang_input, run_button, output_area)
```
