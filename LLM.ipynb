Define Translation Function

```python

def translate_text(text, target_language="French"):
    response = client.chat.completions.create(
        model="gpt-4",
        messages=[
            {"role": "system", "content": f"You are a translator that converts text into {target_language}."},
            {"role": "user", "content": text}
        ]
    )
    return response.choices[0].message.content
```

Define Summarization Function

```python
def summarize_text(text):
    response = client.chat.completions.create(
        model="gpt-4",
        messages=[
            {"role": "system", "content": "You are a summarizer that condenses text into a short summary."},
            {"role": "user", "content": text}
        ]
    )
    return response.choices[0].message.content
```

Cell 4 — Chain Actions (Translation → Summarization)

```python
original_text = "The training process of a PyTorch model involves forward and backward passes."

# Step 1: Translate
translated = translate_text(original_text, target_language="Spanish")
print("Translated Text:", translated)

# Step 2: Summarize
summary = summarize_text(translated)
print("\nSummary of Translation:", summary)
```

Multi‑Language Loop

```python
languages = ["French", "German", "Hindi"]

for lang in languages:
    translated = translate_text("Hello, how are you?", target_language=lang)
    summary = summarize_text(translated)
    print(f"\n--- {lang} ---")
    print("Translation:", translated)
    print("Summary:", summary)
```
