```python
from openai import OpenAI
from pydantic import BaseModel

client = OpenAI()

class Step(BaseModel):
    explanation: str
    output: str

class MathReasoning(BaseModel):
    steps: list[Step]
    final_answer: str

def outputparsingintosteps(math_reasoning):
    # Access the list of steps
    for step in math_reasoning.steps:
        print("Explanation:", step.explanation)
        print("Output:", step.output)
        print("---")
    
    # Access the final answer
    print("Final Answer:", math_reasoning.final_answer)

response = client.responses.parse(
    model="gpt-4o-2024-08-06",
    input=[
        {
            "role": "system",
            "content": "Act as a helpful math tutor. Solve the given problem step by step, ensuring that each step is written on a new line for clarity.",
        },

        {"role": "user", "content": "how can I solve 8x + 7 = -23"},
        
    ],
    text_format=MathReasoning,
)

math_reasoning = response.output_parsed

outputparsingintosteps(math_reasoning)
```
