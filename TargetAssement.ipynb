```python
from openai import OpenAI

# Initialize client (make sure your API key is set in environment variables)
client = OpenAI()

def translate_text(text):
    prompt = f"""
    Extract opinion targets and assessments from the following text.
    Return them as JSON with keys 'target' , 'assessment' ,'intent','utterance','entities'.
    
    Text: "{text}"
    """
    
    response = client.chat.completions.create(
        model="gpt-4o-mini",  # lightweight model, good for structured extraction
        messages=[{"role": "user", "content": prompt}],
        temperature=0  # deterministic output
    )
    print(response.choices[0].message.content)

# Example text
translate_text("The hotel room was excellent, but the service was terrible.")
translate_text("What is the time in London?")
```
