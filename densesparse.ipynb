```python
import openai
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
```

Generate dense embeddings with OpenAI

```python
response = openai.embeddings.create(
    model="text-embedding-3-small",
    input="Sparse embeddings are useful for keyword-based retrieval."
)
dense_vector = response.data[0].embedding
```

Generate sparse embeddings (TF-IDF)

```python
corpus = [
    "Sparse embeddings are useful for keyword-based retrieval.",
    "Dense embeddings capture semantic meaning."
]

vectorizer = TfidfVectorizer()
sparse_matrix = vectorizer.fit_transform(corpus)
sparse_vector = sparse_matrix.toarray()[0]
```

Hybrid scoring (dense + sparse)

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Dense similarity (between two dense vectors)
dense_sim = cosine_similarity([dense_vector], [dense_vector])[0][0]

# Sparse similarity (between two sparse vectors)
sparse_sim = cosine_similarity([sparse_vector], [sparse_vector])[0][0]

# Weighted hybrid score
alpha = 0.5  # balance factor between dense and sparse
hybrid_score = alpha * dense_sim + (1 - alpha) * sparse_sim

print("Dense similarity:", dense_sim)
print("Sparse similarity:", sparse_sim)
print("Hybrid similarity:", hybrid_score)
```
