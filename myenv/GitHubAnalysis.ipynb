```python
from openai import OpenAI
import subprocess

client = OpenAI()

def git_clone(repo_url):
    subprocess.run(["git", "clone", repo_url])

def analyze_code(path):
    response = client.chat.completions.create(
        model="gpt-4.1-mini",
        messages=[{"role": "user", "content": f"Review code in {path}"}]
    )
    return response.choices[0].message.content

# Example usage
git_clone("https://git@github.com:PankajWagh/AI-Automation.git")
print(analyze_code("./main"))
```
