Load Data Using Pandas

```python
import pandas as pd

data = {
"campaign":["Summer Sale","Travel Deals","Holiday Promo"],
"channel":["Facebook","Google Ads","Instagram"],
"impressions":[10000,12000,9000],
"clicks":[800,900,700],
"conversions":[120,150,100],
"cost":[500,700,400]
}

df = pd.DataFrame(data)

print(df)
```

Click Through Rate (CTR)

```python
df["CTR"] = df["clicks"] / df["impressions"]
```

Conversion Rate

```python
df["conversion_rate"] = df["conversions"] / df["clicks"]
```

Cost Per Conversion

```python
df["cost_per_conversion"] = df["cost"] / df["conversions"]
```

Visualization of Campaign Performance

```python
import matplotlib.pyplot as plt

df.plot(x="campaign", y="conversions", kind="bar")

plt.title("Campaign Conversions")
plt.xlabel("Campaign")
plt.ylabel("Conversions")

plt.show()
```

Generate AI Marketing Suggestions

```python
from openai import OpenAI

client = OpenAI()

summary = df.to_string()

response = client.responses.create(
    model="gpt-4.1-mini",
    input=f"Analyze this digital marketing campaign data and suggest improvements:\n{summary}"
)

print(response.output_text)
```
