```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain.chains import LLMChain
import pandas as pd
from openai import OpenAI
from pypdf import PdfReader
```

```python
# Initialize LLM
llm = ChatOpenAI(model="gpt-4-turbo")
```

```python
# Define a prompt template for sentiment analysis
prompt = PromptTemplate(
    input_variables=["text"],
    template="Analyze the sentiment of the following text. "
             "Classify as Positive, Negative, or Neutral, and explain briefly:\n\n{text}"
)
```

```python
client = OpenAI()
```

```python
# Build chain
sentiment_chain = prompt | llm
```

```python
# Example: Document-level
# creating a pdf reader object
reader = PdfReader('comment.pdf')

# printing number of pages in pdf file
print(len(reader.pages))

# getting a specific page from the pdf file
page = reader.pages[0]

# extracting text from page
document = page.extract_text()
print(document)

result_doc = sentiment_chain.invoke(document)
print("Document Sentiment:", result_doc)
```

```python
# Example: Document-level
document = """The new HR policy is very supportive of employees, 
but some feel the leave restrictions are too strict."""
result_doc = sentiment_chain.invoke(document)
print("Document Sentiment:", result_doc) 
```

```python
# Example: Paragraph-level
paragraph = "Excellent smartwatch with great features at a very reasonable price! The display is bright, clear, and responsive, making it easy to use even in sunlight. The design is sleek and comfortable, perfect for all-day wear. Battery life is impressive—I can easily use it for several days on a single charge. The fitness tracking features like heart rate monitoring, step count, and sleep tracking are accurate and very helpful for daily health monitoring. It also connects smoothly with the phone, and notifications come instantly without any lag. Overall, a perfect combination of style, performance, and value for money. Highly recommended for anyone looking for a reliable smartwatch!"
result_para = sentiment_chain.invoke(paragraph)
print("\n Paragraph Sentiment:", result_para)

paragraph1 = "Your first response times have gone up in the last few weeks, so customers are waiting longer than normal to hear back about their issues. Are there any bugs or products that you’re having trouble with? Let’s go over some of your tickets and talk about what aspects might be slowing down your responses."
result_para1 = sentiment_chain.invoke(paragraph1)
print("\n Paragraph Sentiment:", result_para1)


paragraph2 = "Battery draining very fast even without doing anything. Connection & synchronisation problem. Very much lagging. Much better products available at less than its half price."
result_para2 = sentiment_chain.invoke(paragraph2)
print("\n Paragraph Sentiment:", result_para2)
```

```python
data = {
    "review": [
        "The product quality is amazing!",
        "Delivery was very slow",
        "Customer support helped me quickly",
        "Very bad experience",
        "Excellent service and fast shipping",
        "The packaging was damaged",
        "Great product value"
    ]
}

df = pd.DataFrame(data)
```

```python
def get_sentiment(review):
    result_para_sentiment = sentiment_chain.invoke(review)
    print("\n sentence Sentiment:", result_para_sentiment)
```
